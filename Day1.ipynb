```python
# Number guessing game

secret = random.randint(0,100)
tries = 0

while True:
    guess = int(input('Your guess? : '))
    tries += 1
    
    if guess < secret:
        print('Go higher')
    elif guess > secret:
        print('Go lower')
    else:
        print(f'You got it in {tries} tries!')
#         print('You got it!')
#         print('In', tries, 'tries')
#         print('In' + tries + 'tries')
        break
```
